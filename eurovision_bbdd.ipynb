```python
# import libraries

import duckdb     # pip install duckdb

import pandas as pd
pd.options.display.max_rows = None
import os
```

```python
# List the `.csv` data files and create the relative paths list

# List .csv filenames

dir_list = os.listdir('./data/')
dir_list
```

```python
# Create relative paths list (i.e.: tables)

tables = [f'./data/{file}' for file in dir_list]
tables
```

```python
# Create the DDL queries

queries = [f"CREATE OR REPLACE TABLE {table[7:-4]} AS SELECT * FROM '{table}';" for table in tables]
queries
```

```python
# Load the tables into your Database

# Create database connection and .db file

con = duckdb.connect('./data/eurovision_database.db')
```

```python
# Create tables in database and load data from .csv files

for query in queries:
    con.sql(query)
```

```python
country_id_lst = list(con.sql("select * from countries;").df()["country_name"].unique())
country_id_lst
```

```python
country_lst = list(pd.read_html("https://es.wikipedia.org/wiki/Anexo:Participaci%C3%B3n_en_el_Festival_de_la_Canci%C3%B3n_de_Eurovisi%C3%B3n")[0]["País"][:40])
country_lst
```

```python
res = {country_id_lst[i]: country_lst[i] for i in range(len(country_id_lst))}
res
```

```python
countries_df = con.sql("select * from countries;").df()
countries_df
```

```python
df_countries = countries_df.replace(res)
df_countries
```

```python
df_countries.to_csv('./data/countries.csv', index=False)
```
